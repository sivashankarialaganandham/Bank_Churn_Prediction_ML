**Detect Outliers**

```python
import numpy as np

# Define a function to detect outliers using Z-score
def detect_outliers_z_score(data, threshold=3):
    z_scores = np.abs((data - data.mean()) / data.std())
    return z_scores > threshold

# Selecting only numerical columns for outlier detection
numerical_columns = df.select_dtypes(include=np.number)

# Detect outliers using Z-score for each numerical column
outliers = numerical_columns.apply(detect_outliers_z_score)

# Summarize the presence of outliers
outliers_summary = outliers.sum()
print(outliers_summary)
```
